```python
from getpass import getpass
from json import load,dump
import time
f=open('bank.db')
bank=load(f)
f.close()
admin='root'
admin_password='redhat@123'
while True:
    print("\n\n")
    print("*"*50)
    print(time.ctime())
    f=open('log.txt','a')
    s="{} ....Application run by user \n ".format(time.ctime())
    f.write(s)
    f.close()
    print("welcome to SBI Bank ,we care for you")
    print("*"*50)
    print("\n\n1. Login \n2. Signup \n3. Admin \n4. Exit")
    ch=int(input("your choice:  "))
    
    if ch==1:
        print("\n\n")
        uacc=int(input("Account Number:  "))
        if uacc in bank['Account number']:
            upass=getpass("password:  ")
            index=bank['Account number'].index(uacc)
            if upass==bank['Password'][index]:
                s="{}...... User with account number {} logged into account \n".format(time.ctime(),uacc)
                f=open('log.txt','a')
                f.write(s)
                f.close()
                while True:
                    print("Select an option from our menu")
                    print("option 1.: Withdrawl ")
                    print("option 2.: Deposit ")
                    print("option 3.: Check Balance ")
                    print("option 4.: Change Password ")
                    print("option 5.: Logout ")
                    option=int(input("Choose an option"))
                    if option==1:
                        withdrawl=float(input("how much do you want to withdrawl?:"))
                        if withdrawl<=bank['Balance'][index] and withdrawl>0:
                            bank['Balance'][index]=bank['Balance'][index]-withdrawl
                            print('new balance: ',bank['Balance'][index])
                            f=open('log.txt','a')
                            s="{} ....user with accont number {} withdrawl {} rupees\n ".format(time.ctime(),uacc,withdrawl)
                            f.write(s)
                            f.close()
                            f=open('bank.db','w')
                            dump(bank,f)
                            f.close()
                        else:
                            (print("Insufficient balance or \n you entered an invalid amount"))
                    elif option==2:
                        Deposit=float(input("how much do you want to deposit?:"))
                        bank['Balance'][index]=bank['Balance'][index]+Deposit
                        f=open('bank.db','w')
                        dump(bank,f)
                        f.close()
                        print("\n your account successfully updated")
                        print("You have {} rupees in your account :".format(bank['Balance'][index]))
                        f=open('log.txt','a')
                        s="{} ....user with accont number {} Deposit {} rupees\n ".format(time.ctime(),uacc,Deposit)
                        f.write(s)
                        f.close()
                    elif option==3:
                        print('Name: ',bank['User'][index])
                        print('Account number: ',bank['Account number'][index])
                        print('Balance:',bank['Balance'][index])
                    elif option==4:
                        upass = getpass("New password : ")
                        upass1=getpass("Verify password :")
                        if upass==upass1:
                              bank['Password'][index]=upass
                              f=open('bank.db','w')
                              dump(bank,f)
                              f.close()
                              print("Password successfully changed")
                              break
                        else:
                              print("\n Password does not match try again")
                        
                    elif option==5:
                        ch=input("do you really want to logout (y/n) :").strip().lower()
                        if ch=='y' or ch=='yes':
                            print("Thanks for using our services")
                            break
                        else:
                            print("\n Invalid choice")
                            print("Try again")
                    else:
                        f=open('log.txt','a')
                        s="{}.....Wrong Try by user \n".format(time.ctime())
                        f=write(s)
                        f.close()
                        print("\n\n No such Account exists \n You should signup \n\n")
    elif ch==2:
        Acc=bank['Account number'][-1] +1
        name=input("Username : ")
        password=getpass('Set Password : ')
        Balance=float(input('Initial Balance : '))
        bank['User'].append(name)
        bank['Account number'].append(Acc)
        bank['Password'].append(password)
        bank['Balance'].append(Balance)
        f=open('bank.db','w')
        dump(bank,f)
        f.close()
        print("\n Account Successfully created")
        print('Note down your account number: {}'.format(Acc))
        print('Now You can Login\n')
    elif ch==3:
        Username=input('\n Enter name  : ').strip().lower()
        password=getpass()
        if Username == admin:
            if password== admin_password:
                print()
                for key,value in bank.items():
                    print("{}={}".format(key,value))
                print()
            else :
                print("\nInvalid password \nTry Again")
        else :
            print("\nThis user is not an admin.")
    elif ch == 4 :
        ch = input("\n\nDo really want to exit ? (y/n) : ").strip().lower()
        if ch == 'y' or ch == 'yes' :      
            print("\nThanks for useing our services ")
            break
    else :
        print("\nInvalidChoice\nTry again")
    
```
